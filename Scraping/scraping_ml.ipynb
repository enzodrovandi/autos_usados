```python
def get_info(url):
    r = requests.get(url)
    soup = bs(r.content)
    contents = soup.prettify()
    
    main = soup.find('main')
    title = main.find('h1', class_= 'ui-pdp-title').get_text()
    price = main.find(class_='andes-money-amount__fraction').get_text().replace(".","")
    currency_symbol = main.find(class_='andes-money-amount__currency-symbol').get_text()
    info_table = main.find(class_= 'andes-table')
    info_rows  = info_table.find_all('tr')
    
    car_info = {}
    
    for idx, row in enumerate(info_rows):
        car_info['url'] = url
        car_info['title'] = title
        car_info['price'] = price
        car_info['currency_symbol'] = currency_symbol
        content_key = row.find('th').get_text(strip=True)
        content_value = row.find('td').get_text(strip=True)
        car_info[content_key] = content_value
    
    return car_info
```

### Cambiar dtypes


```python
df = pd.read_csv('autos_usados.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
# Cambiamos el nombre de las columnas para que no contengan acentos

df.rename(columns={'Kil√≥metros':'km', 'Tipo de carrocer√≠a': 'carroceria', 'Motor': 'motor', 'Transmisi√≥n': 'transmision', 'Puertas': 'puertas', 'Tipo de combustible': 'combustible',
          'Color': 'color', 'A√±o': 'year', 'Modelo':'modelo', 'Marca':'marca'}, inplace=True)
```

```python
# Convertimos todos los valores a minuscula
df = df.apply(lambda x: x.str.lower() if(x.dtype == 'object') else x)
```

```python
df['currency_symbol'] = df['currency_symbol'].astype('category')
df['marca'] = df['marca'].astype('category')
df['modelo'] = df['modelo'].astype('category')
df['color'] = df['color'].astype('category')
df['combustible'] = df['combustible'].astype('category')
df['transmision'] = df['transmision'].astype('category')
df['carroceria'] = df['carroceria'].astype('category')
```

```python
df['km'] = df['km'].str.replace('km', "")
```

```python
df['km'] = df['km'].astype('int')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Rellenamosn nulos de color por la moda que es el blanco
df.color.fillna(df.color.mode()[0], inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
#Convertimos todos los precios en dolares a pesos

dolar_blue = 200

df['price_in_pesos'] = df.apply(lambda row: row.price * dolar_blue if row.currency_symbol == 'u$s' else row.price, axis=1)
```
