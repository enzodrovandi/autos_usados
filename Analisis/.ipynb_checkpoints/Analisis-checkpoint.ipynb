```python
import regex as re
```

```python
df2.motor.unique()
```

```python
df2.motor = df2['motor'].str.replace(r'^(?![1-9][.,][0-9]).*$','', regex=True)
```

```python
regex = r'([1-9][.,][0-9])'
df2['motor']=df2.motor.str.extract(regex)
```

```python
df2.motor.unique()
```

```python
fig, axes = plt.subplots(5, 2, figsize=(20,30))
axe = axes.ravel()


for i, col in enumerate(df2.columns.values[:-1]):
    df2.plot(x=col, y='price_in_pesos', kind='scatter', ax=axe[i])


axes[0,0].set_xticklabels(df2.marca.values, rotation=90)
axes[0,1].set_xticklabels(df2.modelo.values, rotation=30)
axes[1,1].set_xticklabels(df2.color.values, rotation=90)
axes[3,1].set_xticklabels(df2.motor.values, rotation=90)

fig.tight_layout()
plt.show()
```
