```python
fig, axes = plt.subplots(5, 2, figsize=(20,30))
axe = axes.ravel()


for i, col in enumerate(df2.columns.values[:-1]):
    df2.plot(x=col, y='price_in_pesos', kind='scatter', ax=axe[i])


axes[0,0].set_xticklabels(df2.marca.unique(), rotation=90)
axes[1,0].set_xticklabels(df2.color.unique(), rotation=90)
axes[3,0].set_xticklabels(sorted(df2.motor.unique()), rotation=90)
axes[3,1].set_xticklabels(df2.carroceria.unique(), rotation=90)

fig.tight_layout()
plt.show()
```
